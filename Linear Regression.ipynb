```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class LinearRegression(object):
    
    def __init__(self, eta, n_iter):               #eta is learning hyperparameter alpha and n_iter is no of iterations
        self.eta = eta
        self.n_iter = n_iter
    
    def fit(self, x, y):                           #this function is to train the linear regression model
        self.cost = []
        self.para = np.zeros((x.shape[1],1))
        m = x.shape[0]
        
        for _ in range(self.n_iter):
            y_pred = np.dot(x,self.para)
            diff = y_pred - y
            gradient_vec = np.dot(x.T,diff)
            self.para -= (self.eta/m)*gradient_vec
            temp_cost = np.sum(np.square(diff))/(2*m)
            self.cost.append(temp_cost)
        return self
    
    def predict(self, x):                          #this function is to predict the output
        return np.dot(x,self.para)
```

```python
np.random.seed(0)                               #generating dataset 
x = np.random.rand(100,1)
y = 16 + 3*x + np.random.rand(100,1)
```

```python
plt.scatter(x,y,s = 20, c = 'blue')             #printing the dataset
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
model = LinearRegression(0.05,10)          #created class with alpha and no of iterations as parameters
add = lambda r:r+5                           #5 is the bias that we have added 
x_train = add(x)                             
model = model.fit(x_train,y)                 #training
y_p = model.predict(x_train)                 #predicting

plt.scatter(x,y,s = 20, c = 'blue')          #plotting the dataset
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x,y_p,c = 'red')                    #ploting the output
plt.show()
```

```python
plt.plot(model.cost)                        #graph for cost function over 10 iterations
plt.show()
```
